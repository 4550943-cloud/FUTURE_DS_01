```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import warnings
warnings.filterwarnings('ignore') # Hides annoying warning messages

# Set overall style
sns.set_theme(style="whitegrid")

# Load the main Superstore data
df_main = pd.read_csv('/content/Sample - Superstore.csv', encoding='latin1') # Assuming latin1 encoding and standard Superstore columns

# 1. Prepare Region Data
df_region = df_main.groupby('Region')[['Sales', 'Profit']].sum().reset_index()
df_region = df_region.rename(columns={'Region': 'Row Labels', 'Sales': 'Sum of Sales', 'Profit': 'Sum of Profit'})

# 2. Prepare Category Data
df_category = df_main.groupby('Category')['Sales'].sum().reset_index()
df_category = df_category.rename(columns={'Category': 'Row Labels', 'Sales': 'Sum of Sales'})

# 3. Prepare Cleaned Data for Sub-Category Profit and Time Trend
df_clean = df_main.copy()
df_clean['Order Date'] = pd.to_datetime(df_clean['Order Date'], format='%m/%d/%Y') # Assuming MM/DD/YYYY format for 'Order Date'

# Group by Sub-Category to find Top and Bottom performers
df_subcat = df_clean.groupby('Sub-Category')['Profit'].sum().reset_index()
df_subcat = df_subcat.sort_values(by='Profit', ascending=False)
top_5 = df_subcat.head(5)
bottom_5 = df_subcat.tail(5)
df_top_bottom = pd.concat([top_5, bottom_5])

# Group by Month for Time Trend
df_time = df_clean.set_index('Order Date').resample('M')['Sales'].sum().reset_index()

# Create a 2x2 subplot figure
fig, axes = plt.subplots(2, 2, figsize=(16, 12))
fig.suptitle('Business Sales Performance Dashboard', fontsize=20, fontweight='bold')

# --- Plot 1: Sales and Profit by Region (Top Left) ---
x = np.arange(len(df_region['Row Labels']))
width = 0.35
axes[0, 0].bar(x - width/2, df_region['Sum of Sales'], width, label='Sales', color='skyblue')
axes[0, 0].bar(x + width/2, df_region['Sum of Profit'], width, label='Profit', color='salmon')
axes[0, 0].set_ylabel('Amount ($)')
axes[0, 0].set_title('Sales and Profit by Region', fontsize=14)
axes[0, 0].set_xticks(x)
axes[0, 0].set_xticklabels(df_region['Row Labels'])
axes[0, 0].legend()

# --- Plot 2: Sales by Category (Top Right) ---
axes[0, 1].pie(df_category['Sum of Sales'], labels=df_category['Row Labels'], autopct='%1.1f%%',
               startangle=90, colors=['#ff9999','#66b3ff','#99ff99'])
axes[0, 1].set_title('Sales Distribution by Category', fontsize=14)

# --- Plot 3: Top & Bottom 5 Sub-Categories by Profit (Bottom Left) ---
colors = ['green' if p > 0 else 'red' for p in df_top_bottom['Profit']]
axes[1, 0].barh(df_top_bottom['Sub-Category'], df_top_bottom['Profit'], color=colors)
axes[1, 0].set_xlabel('Profit ($)')
axes[1, 0].set_title('Top 5 and Bottom 5 Sub-Categories by Profit', fontsize=14)
axes[1, 0].invert_yaxis() # Highest profit at the top

# --- Plot 4: Sales Trend Over Time (Bottom Right) ---
axes[1, 1].plot(df_time['Order Date'], df_time['Sales'], marker='o', linestyle='-', color='purple')
axes[1, 1].set_ylabel('Total Sales ($)')
axes[1, 1].set_title('Monthly Sales Trend (2014 - 2017)', fontsize=14)
axes[1, 1].tick_params(axis='x', rotation=45)

# Save the dashboard as an image
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.savefig('Sales_Dashboard.png')
print("Dashboard successfully saved as Sales_Dashboard.png!")
```
